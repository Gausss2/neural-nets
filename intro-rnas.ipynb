**Rápida Introdução**

**IA**: é um ramo de pesquisa dentro da ciência da computação que procura simular a inteligência humana em máquinas, em outras palavras, programar uma máquina para pensar e tormar decisões como um ser humano.

Repare que todas as decisões são tomadas com base em uma informação, essa é a peça chave do machine learning, de onde é que o o algorítmo irá extrair essa informação.

**Exemplo de Problema**


Vamos supor que este neurônio é capaz de classificar se uma paciente é ou não portador de coronavirus. nosso dado de entrada será essa tabela (vetor):

**Entradas**

<table>
  <tr> 
    <th> Tem Febre?</th> <th>Tem Tosse?</th> <th>Tem Falta de Ar?</th>
   </tr>
  
  <tr>
      <td> 1 </td> <td>1 </td> <td> 0 </td>
  </tr>
</table> 

<i>1 corresponde a "sim" e 0 a "não".</i>

Cada valor dessa tabela será passado ao neurônio da seguinte forma:

>$x_{1}=1$

>$x_{2}=1$

>$x_{3}=0$

**Pesos**

Pesos são valores que definem o quão importante uma entrada é para a tomada de decisão. São representados por **$w$** e cada entrada possui seu respectivo peso. 

Essa definição de importância das entradas é feita a partir do produto entre a entrada e seu peso.

Vamos supor também que Febre é o sintoma mais relevante para o diagnóstico de coronavirus, tosse o segundo e por último falta de ar.

>$w_{1}=0.83$

>$w_{2}=0.67$

>$w_{3}=0.21$

*Obs.: Os valores de pesos costumam ficar entre 0.00 e 1.00, quanto mais próximo de 1 o valor do peso estiver mais relevante a respectiva entrada será e quanto mais próximo de 0 menos relevante.*




A nossas possíveis saídas são:

>$y=1$
 se o paciente for portador e $y=0$ se o paciente não for portador.

**Função de Ativação e Bias**

Função de ativação é responsável por por decidir se o neurônio será disparado, ou seja, a saída ser 1.

Porém, como toda função ela só possui uma entrada e uma saída, sendo assim é necessário transformar todas as entradas em um único valor... Aí que entra a **soma ponderada**, que consiste na seguinte equação:

>$u=(x_{1}*w_{1}) + (x_{2}*w_{2}) + (x_{3}*w_{3})$
ou
>$u=\displaystyle\sum_{i=1}^3 (x_{i}*w_{i})$


há também o paraâmetro não mostrado, mas que está bastante presente em redes neurais o **<i>bias</i>**, bias é uma constante que ajuda o modelo a se adaptar 
a diferentes dados, ampliando seu grau de conhecimento (geralmente possui valor 1 ou -1). 


(isso ficará mais claro quando for apresentado o treinamento das redes neurais, mas por enquanto vamos apenas aceitá-lo.)


Com bias a equação ficará da seguinte forma:
>$u=(x_{1}*w_{1}) + (x_{2}*w_{2}) + (x_{3}*w_{3}) + b$
ou
>$u=\displaystyle\sum_{i=1}^3 (x_{i}*w_{i}) + b$

Com a entrada, agora aplica-se a seguinte função de ativação... Existem dezenas, mas a mais introdutório chama-se relu que consiste na seguinte equação:


$f(u)=$\begin{cases}
    1       & \quad \text{if } u \text{ > 0}\\
    0  & \quad \text{if } u \leq\text{0}
\end{cases}

ou seja, para todo $u$ positivo $y=1$ e para todo $u$ negativo $y=0$.

**Aplicando o neurônio**

Esse é funcionamento básico de um neurônio perceptron. Agora, aplicando todas essas equações temos...

>$u=(1*0.83) + (1*0.67) + (0*0.21) - 1 = 0.5$

>$f(2.5) = 1$
ou
>$y=1$

Isso quer dizer que o nosso neurônio determinou nosso paciente como portador de 
corona virus. 

Repare que caso o paciente tivesse apenas tivesse um sintoma menos relevante como Falta de ar o resultado seria diferente.

**No pior caso**:

  <tr>

   </tr>
  
  <tr>

  </tr>

>$u=(1*0.83) + (1*0.67) + (1*0.21) - 1 = 0.71$

>$f(0.71) = 1$
ou
>$y=1$



**No melhor caso**:

<table>
  <tr> 
    <th> Tem Febre?</th> <th>Tem Tosse?</th> <th>Tem Falta de Ar?</th>
   </tr>
  
  <tr>
      <td> 0 </td> <td>0 </td> <td> 0 </td>
  </tr>
</table> 

teriamos:

>$u=(0*0.83) + (0*0.67) + (0*0.21) - 1 = -1$

>$f(-1) = 0$
ou
>$y=0$
